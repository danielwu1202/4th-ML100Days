```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
numeric_columns = app_train.select_dtypes(['float64','int64'])
dtype_select = ['float64','int64']
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns" ,len(numeric_columns))

for column in numeric_columns:
    print(column)
    plt.figure(figsize=(5,3))
    app_train[column].hist(bins=40)
    plt.title(column)
    plt.show()
```

```python
print(app_train['AMT_INCOME_TOTAL'].describe())
```

```python
def ecdf(data):
    xs = np.sort(data) #ç´¯ç©å‡½æ•¸ï¼Œæ¯å€‹è³‡æ–™å¿…è¦é †åºæŽ’åˆ—
    ys = np.arange(1, len(xs)+1)/float(len(xs)) #æ¯å€‹è³‡æ–™å æ¯”
    return xs, ys
```

```python
xs, ys = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.figure(figsize=(14,4))
plt.plot(xs, ys,label="AMT_INCOME_TOTAL")
plt.title('ECDF of AMT_INCOME_TOTAL')
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([xs.min(), xs.max()]) # é™åˆ¶é¡¯ç¤ºåœ–ç‰‡çš„ç¯„åœ
plt.ylim([-0.05,1.05]) # é™åˆ¶é¡¯ç¤ºåœ–ç‰‡çš„ç¯„åœ
plt.legend()
plt.show()

plt.figure(figsize=(14,4))
plt.plot(np.log(xs), ys,label="AMT_INCOME_TOTAL")
plt.title('ECDF of AMT_INCOME_TOTAL')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # é™åˆ¶é¡¯ç¤ºåœ–ç‰‡çš„ç¯„åœ
plt.legend()
plt.show()
```

```python
# æœ€å¤§å€¼è½åœ¨åˆ†å¸ƒä¹‹å¤–
print(app_train['REGION_POPULATION_RELATIVE'].describe())
xs, ys = ecdf(app_train['REGION_POPULATION_RELATIVE'])
plt.figure(figsize = (14, 4) )
plt.plot(xs, ys, label = "REGION_POPULAION_RELATIVE")
plt.title("ECDF of REGION_POPULATION_RELATIVE")
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([xs.min(), xs.max()]) # é™åˆ¶é¡¯ç¤ºåœ–ç‰‡çš„ç¯„åœ
plt.ylim([-0.05,1.05]) # é™åˆ¶é¡¯ç¤ºåœ–ç‰‡çš„ç¯„åœ
plt.legend()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()
```
