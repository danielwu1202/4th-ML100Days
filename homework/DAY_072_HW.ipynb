```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

```python
x = np.arange(-10, 10, 0.01)

y = list(map(lambda x: x if x>0 else 0, x))
plt.plot(x, y, 'r')
plt.grid()
plt.show()

dy = list(map(lambda x: 1 if x>0 else 0, x))
plt.plot(x, dy, 'b')
plt.grid()
plt.show()
```
